```python
import pandas as pd 
import numpy as np 
from sklearn.cluster import k_means
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("AVStream.csv")
vs = []
vsl = []

for val in data['valence stream'][:]:
	try:
		to_num = val[1:-1].split(',')
		#print vs
		vs.append(map(lambda x:float(x.replace(' ','')),to_num))
		vsl.append(len(to_num))
	except:
		print val[1:-1].split(','),'error'



maxl = max(vsl)
#print maxl
ind = 0
for v in vs:
	#print v
	al = [-1]*(maxl - len(v))
	v = v + al
	vs[ind] = v
	ind += 1



# #print vs[0]
#print vs
```

```python
AS = []
asl = []

for val in data['arousal stream'][:]:
	try:
		to_num = val[1:-1].split(',')
		#print vs
		AS.append(map(lambda x:float(x.replace(' ','')),to_num))
		asl.append(len(to_num))
	except:
		print val[1:-1].split(','),'error'



maxl = max(asl)
#print maxl
ind = 0
for v in AS:
	#print v
	al = [-1]*(maxl - len(v))
	v = v + al
	AS[ind] = v
	ind += 1
```
