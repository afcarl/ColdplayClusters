```python
import os, sys
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataText = pd.read_csv('ColdplaySongs.csv')
names = map(lambda x: x[:-4], dataText['songs'])
sel = 20
```

```python
from itertools import combinations

nodes = range(sel)
edges = combinations(nodes, 2)
G = nx.Graph()
G.add_nodes_from(nodes)
G.add_edges_from(edges)#,weight = dataSim.iloc[edges[0][edges[1]])
nx.draw(G)
```

```python
mapping = dict(zip(nodes,names[:sel]))
```

```python
g = nx.Graph()
nodes = range(sel)
G.add_nodes_from(nodes)
for (u, v) in G.edges():
    g.add_edge(u,v,weight=dataSim.iloc[u][v])
```

```python
H=nx.relabel_nodes(g,mapping)
```

```python
nx.draw(H,pos=nx.spring_layout(H),alpha=0.8, style = 'dashed',with_labels = True)
```
